```
class MyListener(StreamListener):

  def on_data(self, data):
    try:
      with open('Scraping_Twitter.csv', 'a') as f:
        f.write(data)
        return True
    except BaseException as e:
      print("Error en el dato: %s" % str(e))
    return True

  def on_error(self, status):
    print(status)
    return True
```
